```python
import google.generativeai as genai
from PIL import Image
import io
from google.colab import files  # Para facilitar o upload de arquivos no Colab
```

```python
def upload_imagens():
  """Permite o upload de múltiplas imagens pelo usuário no Colab."""
  uploaded = files.upload()
  return uploaded
```

```python
def carregar_e_preparar_imagem(nome_arquivo, dados_arquivo):
  """Carrega uma imagem a partir dos dados do arquivo e a prepara para o Gemini."""
  try:
    image = Image.open(io.BytesIO(dados_arquivo))
    # Podemos adicionar aqui alguma lógica de pré-processamento se necessário,
    # como redimensionar a imagem para um tamanho específico, mas por enquanto vamos manter simples.
    return image
  except Exception as e:
    print(f"Erro ao carregar a imagem {nome_arquivo}: {e}")
    return None
```

```python
imagens_carregadas = upload_imagens()
imagens_pil = {}
if imagens_carregadas:
  for nome_arquivo, dados_arquivo in imagens_carregadas.items():
    imagem_pil = carregar_e_preparar_imagem(nome_arquivo, dados_arquivo)
    if imagem_pil:
      imagens_pil[nome_arquivo] = imagem_pil
  print(f"\nImagens carregadas e preparadas: {list(imagens_pil.keys())}")
else:
  print("\nNenhuma imagem foi carregada.")
```

```python
def analisar_pecas_e_caracteristicas(imagem):
  """Analisa as peças de roupa em uma imagem, descreve suas características e sugere nomes."""
  prompt_pecas = """Você é um especialista em moda. Analise a imagem e seja sucinto.
  1. Identifique cada peça de roupa presente na imagem.
  2. Para cada peça, descreva suas características principais, como cor, tipo de tecido (se possível identificar), modelagem, detalhes (estampas, bordados, etc.).
  3. Sugira um nome comum para cada peça identificada.

  Formate sua resposta de forma clara e concisa, listando cada peça com suas características e nome sugerido."""

  response = model.generate_content([prompt_pecas, imagem])
  return response.text

def analisar_ocasiao_e_combinacoes(analise_pecas, imagem):
  """Analisa a ocasião mais adequada para o look e sugere possíveis combinações."""
  prompt_ocasiao = f"""Com base na seguinte análise das peças de roupa:
  '{analise_pecas}'
  e considerando a imagem da roupa, determine:
  1. Para qual tipo de ocasião ou evento este look seria mais adequado (ex: casual, trabalho, festa, dia a dia, etc.)? Justifique sua resposta com base nas características das peças.
  2. Sugira pelo menos duas possíveis combinações adicionais de peças (que não estão necessariamente na imagem) que complementariam este look para a ocasião identificada. Seja específico sobre as peças e suas características (cor, estilo, etc.)."""

  response = model.generate_content([prompt_ocasiao, imagem])
  return response.text
```

```python
resultados_analise = {}
if imagens_pil:
  for nome_arquivo, imagem in imagens_pil.items():
    print(f"\n--- Analisando a imagem: {nome_arquivo} ---")

    # Agente 1: Análise de peças e características
    analise_pecas = analisar_pecas_e_caracteristicas(imagem)
    print("\nAnálise das Peças e Características:")
    print(analise_pecas)

    # Agente 2: Análise de ocasião e combinações
    analise_ocasiao = analisar_ocasiao_e_combinacoes(analise_pecas, imagem)
    print("\nAnálise de Ocasião e Possíveis Combinações:")
    print(analise_ocasiao)

    resultados_analise[nome_arquivo] = {
        "analise_pecas": analise_pecas,
        "analise_ocasiao": analise_ocasiao
    }
else:
  print("Nenhuma imagem para analisar.")

# (Opcional) Podemos imprimir um resumo dos resultados ao final
if resultados_analise:
  print("\n--- Resumo da Análise ---")
  for nome_arquivo, resultados in resultados_analise.items():
    print(f"\nImagem: {nome_arquivo}")
    print(f"  Análise das Peças: {resultados['analise_pecas']}")
    print(f"  Análise de Ocasião e Combinações: {resultados['analise_ocasiao']}")
```
